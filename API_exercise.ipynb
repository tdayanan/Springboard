```python
# Task 1 : Collect data from the Franfurt Stock Exchange, for the ticker AFX_X, for the whole year 2017 (keep in mind that the date format is YYYY-MM-DD)


#Collect data from the Frankfurt Stock Exchange, for the ticker AFX_X for the whole year 2017
year_2017 = [info for info in r.json()['dataset']['data'] if info[0][:4]=='2017']

print(year_2017)

```

```python
# Task 2 : Convert the returned JSON object into a Python dictionary.¶


#First, get the columns names to use it as keys to store data in them
columns = r.json()['dataset']['column_names']

from collections import OrderedDict

#initialize an empty Ordered dictionary to save a dictionary with nested dictionary
#outer dictionary acts a index for a pandas dataframe
#inside dictionary stores the data i.e columns and rows
d = OrderedDict()
for rows in year_2017:
    new_dict = OrderedDict(zip(columns, rows))
    d[rows[0]] = new_dict
print(d)


```

```python
#task 3: Calculate what the highest and lowest opening prices were for the stock in this period


#create a list to store the Opening prices for the stock
#There are some None type values which we need to take care of.
open_list = [d[dt]['Open'] for dt in d.keys() if d[dt]['Open'] is not None]

```

```python
# TO find highest 
max(open_list)
```

```python
#To find lowest

min(open_list)
```

```python
#task 4  What was the largest change in any one day (based on High and Low price)

change = []
#use for loop to iterate over each row/records
for dt in d.keys():
    #calculate the change in price based in High and Low price
    diff = d[dt]['High']-d[dt]['Low']
    #add the change in price to the list change
    change.append(diff)
```

```python
# the largest change in any one day

max(change)

```

```python
#task 5: What was the largest change between any two days (based on Closing Price)?

#create a list named closing_price to hold the Closing price for the whole year
closing_price = [d[dt]['Close'] for dt in d.keys() if d[dt]['Close'] is not None]


#create another empty list named change_closing_price to hold the change of closing price between two days
change_closing_price = []
#use the for loop to iterate over all the closing prices
for num in range(len(closing_price)):
    #we will only find the difference up to the last item in the list
    if num < (len(closing_price)-1):
        diff = closing_price[num] - closing_price[num+1]
        #The difference may be positive or negative, so we will use the absolute value for the change
        change_closing_price.append(abs(diff))
        

#The largest change between any two days based on Closing Price #The lar 
max(change_closing_price)
```

```python
#task 6:  What was the average daily trading volume during this year?

#Create an empty list named traded_volume to hold the 'Traded Volume' data for the whole year
traded_volume =[d[dt]['Traded Volume'] for dt in d.keys()]


#Calculate the average daily trading volume #Calcula 
average_trading_volume = sum(traded_volume)/len(traded_volume)
average_trading_volume

```
