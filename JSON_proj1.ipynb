```python
#JSON Exercise 2. Find the top 10 major project themes (using column 'mjtheme_namecode')


#print(df.mjtheme_namecode)

from pandas.io.json import json_normalize
json_data=json.load(open("world_bank_projects.json"))
theme_df=json_normalize(json_data,'mjtheme_namecode')
print(theme_df)
# We can get the top 10 major project themes using the value_counts() on name of the theme
print (theme_df.name.value_counts().head(10))


#data=df.to_json()
#new_list=dict(df.mjtheme_namecode)
#json_normalize(new_list)

```

```python
#JSON Exercise 3. In 2. above you will notice that some entries have only the code and the name is missing. 
#Create a dataframe with the missing names filled in.

#Removing the null values to create lookup table with unique values
theme_df1 = theme_df[theme_df.name != ""]
lookup_df= theme_df1.drop_duplicates().sort_values(by='code')
print(lookup_df)

#Merging with left join to fill missing names
merged_df= theme_df.merge(lookup_df,on='code',how='left')

print(merged_df)

print(merged_df.name_y.value_counts())
```
